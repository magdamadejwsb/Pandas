```python
import pandas as pd
```

# Obiekt Series

```python
# tworzenie obiektu Series z listy
s1 = pd.Series([5, 10, 15, 20])
print(s1)
```

```python
# tworzenie obiektu Series z listy z własnym indeksem
s2 = pd.Series([100, 200, 300], index=["A", "B", "C"])
print(s2)
```

```python
# tworzenie obiektu Series ze słownika
# Klucze słownika → indeks
# Wartości → dane
s3 = pd.Series({"Polska": 38, "Niemcy": 83, "Francja": 65})
print(s3)
```

```python
# tworzenie obiektu Series z tablicy NumPy
import numpy as np

arr = np.array([1.5, 2.5, 3.5])
s4 = pd.Series(arr)
print(s4)
```

```python
# tworzenie obiektu Series z danymi tekstowymi
s5 = pd.Series(["Python", "Pandas", "NumPy"])
print(s5)
```
