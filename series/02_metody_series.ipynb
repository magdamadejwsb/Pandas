```python
import pandas as pd
import numpy as np
```

# Metody obiektu Series

```python
s = pd.Series([10, 20, np.nan, 20, 5], index=list('abcde'))
print(s)
```

## Metody statystyczne

```python
# metoda sum()
print(s.sum())
```

```python
# metoda mean()
print(s.mean())
```

```python
# metoda min() i max()
print(s.min())
print(s.max())
```

```python
# metoda count()
print(s.count())
```

```python
# metoda std()
print(s.std())
```

## Metody opisowe

```python
# metoda describe()
print(s.describe())
```

## Metody do pracy z brakami danych i duplikatami

```python
# metoda dropna()
print(s.dropna())
```

```python
# metoda fillna()
print(s.fillna(0))
```

```python
# metoda drop_duplicates()
print(s.drop_duplicates())
```
